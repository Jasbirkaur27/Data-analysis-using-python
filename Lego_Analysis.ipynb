Showing the number of LEGO sets released year-on-year. How do the number of sets released in 1955 compare to the number of sets released in 2019?

The number of LEGO releases on a line chart using Matplotlib. <br>
<br>

```python
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
```


Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
themes=pd.read_csv("./data/themes.csv")
themes.head(5)
```

```python
themes.isnull().count()
```

```python
themes_by_year=sets.groupby('year').agg({'theme_id':pd.Series.nunique})
```

```python
themes_by_year.rename(columns={'theme_id':'nr_themes'},inplace=True)
themes_by_year.head()
themes_by_year.tail()
```

Plot the number of themes released by year on a line chart. Only included the full calendar years (i.e., exclude 2020 and 2021). 

```python
plt.plot(themes_by_year.index[:-2],themes_by_year.nr_themes[:-2])
```

### Line Charts with Two Seperate Axes

```python
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2],color='Green')
ax2.plot(themes_by_year.index[:-2],themes_by_year.nr_themes[:-2])

ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Sets')
ax2.set_ylabel('Number of Themes')
```

Average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

### Merging (i.e., Combining) DataFrames based on a Key


```python
set_theme_count=sets["theme_id"].value_counts()
set_theme_count[:5]
```

```python
set_theme_count=pd.DataFrame({'id':set_theme_count.index,
                             'set_count':set_theme_count.values})
set_theme_count.head()
```

```python
merged_df = pd.merge(set_theme_count, themes, on='id')
merged_df[:3]
```

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('nr of sets',fontsize=14)
plt.xlabel('Theme Name',fontsize=14)
plt.bar(merged_df.name[:10],merged_df.set_count[:10])
```
